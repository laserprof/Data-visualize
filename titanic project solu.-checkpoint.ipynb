```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
df=sns.load_dataset('titanic')

df
```

```python
df.describe()
```

```python
sns.distplot(df['survived'],kde=False , color='m',bins=30)
```

unfortinatly people who dead more than who lived

```python
sns.distplot(df['age'],kde=False , color='m',bins=30)
```

from age analysis we notice that most peopel in the ship are youngers between 20 and 32

```python
sns.distplot(df['fare'],kde=False , color='m',bins=30)
```

most fare under 100

```python
sns.jointplot(y='survived', x='age', data=df)
```

WE notice that people who survived have more lower age  than people who dead
we also notice that the only 80 years old person in the ship is dead


```python
sns.countplot(x='sex', data=df, palette='mako')
```

```python
sns.countplot(x='class', data=df, palette='mako', hue='sex')
```

more people took third-class ticket and male more than female as we noticed before .

```python
sns.countplot(x='who', data=df, palette='mako')
```

```python
sns.countplot(x='embark_town', data=df, palette='mako', hue='sex')
```

Southampton city has the biggest number

```python
sns.countplot(x='alive', data=df, palette='mako', hue='sex')
```

```python
sns.countplot(x='alone', data=df, palette='mako', hue='sex')
```

```python
sns.boxplot(x='sex', y='age', data=df, hue='alive', palette='mako')
```

```python
sns.boxplot(x='alive', y='age', data=df, hue='sex', palette='mako')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle('2 Box Plots')
sns.boxplot(ax=axes[0], x='survived', y='age', data=df, palette='mako', hue='sex')
sns.boxplot(ax=axes[1], x='age', y='class', data=df, palette='mako', hue='sex')
```

```python
sns.swarmplot(x='survived', y='age', hue='sex', data=df, palette='rocket', dodge=True)
```
